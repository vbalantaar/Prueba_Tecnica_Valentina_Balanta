**Modelo de predicción de datos**
¿Se debe hacer mantenimiento?

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Calculate diff_horas
df_master['diff_horas'] = (df_master['hasta'] - df_master['desde']).dt.total_seconds() / 3600


# Usar solo columnas contextuales
df_model = df_master[["subjuicio", "diff_horas", "state", "city"]].dropna()

# Codificar categóricas
for col in ["state", "city"]:
    le = LabelEncoder()
    df_model[col] = le.fit_transform(df_model[col].astype(str))

# Definir variables
X = df_model.drop("subjuicio", axis=1)
y = df_model["subjuicio"].astype(int)  # 1=pasa, 0=falla

# Split train/test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Modelo
clf = RandomForestClassifier(n_estimators=200, random_state=42, class_weight="balanced")
clf.fit(X_train, y_train)

# Predicciones
y_pred = clf.predict(X_test)
print("Reporte de clasificación:\n", classification_report(y_test, y_pred))

# Probabilidades
probs = clf.predict_proba(X_test)[:, 0]  # probabilidad de falla (clase 0)

# Crear recomendación de mantenimiento
threshold = 0.6  # umbral configurable
mantenimiento = (probs > threshold).astype(int)  # 1=requiere mantenimiento

# Mostrar primeras recomendaciones
df_resultados = X_test.copy()
df_resultados["prob_falla"] = probs
df_resultados["requiere_mantenimiento"] = mantenimiento
print(df_resultados.head(10))
```

Accuracy aprox 69% = El modelo acierta casi 7 de cada 10 veces.

**Es decir, el modelo identifica bastante bien cuándo un medidor va a fallar.**

**Aquí el modelo es menos preciso, tiende a dar algunos falsos positivos (decir que fallará cuando en realidad pasa).**
